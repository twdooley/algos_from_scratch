```python
def euc_(p1, p2):
    dists = 0.0
    for row in range(len(p1)):
        dists += np.linalg.norm(p1[row] - p2[row])
    return math.sqrt(dists)
```

### Build a Manhattan Distance function

```python
def man_(p1, p2):
    dists = 0.0
    for row in range(len(p1)-1):
        dists += abs(p1[row] - p2[row])**2
    return math.sqrt(dists)
```

```python
class NearestNeighbors:
    """Class to run k-NN algo.
    .fit(X_train, y_train): sets the plane with the training data. Stores labels. 
    .predict(x_test): predicts the class based on the nearest neighbors. """

    def __init__(self, k):
        self.k = k

    def fit(self, X_train, y_train):
        self.train = X_train
        self.labels = y_train
        #return train, labels

    def euc_(p1, p2):
        dists = 0.0
        for row in range(len(p1)):
            dists += np.linalg.norm(p1[row] - p2[row])
        return math.sqrt(dists)

    def predict(self, x_test):
        train_lab = zip(self.train, self.labels)
        dists = []
        for X_row in train_lab:
            dist = euc_(x_test, X_row[0:2])
            dists.append((X_row, dist))
        dists.sort(key = lambda x: x[1])
        nearestn = []
        for i in range(self.k):
            nearestn.append(dists[i])
        self.nearestn = nearestn
        one_val = nearestn[0][0][1]
        all_val = [row[0][1] for row in nearestn]
        preds = max(set(all_val), key = all_val.count)
        return preds, all_val, nearestn
```
