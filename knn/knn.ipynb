### Build a Euclidean Distance function

```python
# def euc_(p1, p2):
#     dists = 0.0
#     for row in range(len(p1)-1):
#         dists += (p1[row] - p2[row])**2
#     return math.sqrt(dists)
```

```python
def euc_(p1, p2):
    dists = 0.0
    for row in range(len(p1)):
        dists += np.linalg.norm(p1[row] - p2[row])
    return math.sqrt(dists)
```

### Build a Manhattan Distance function

```python
def man_(p1, p2):
    dists = 0.0
    for row in range(len(p1)-1):
        dists += abs(p1[row] - p2[row])**2
    return math.sqrt(dists)
```

```python
def get_knn(X, y):
    dists = []
    for X_row in X:
        dist = euc_(y, X_row)
        dists.append((X_row, dist))
    return(dists)
```

```python
def get_knn(X, y, k):
    dists = []
    for X_row in X:
        dist = euc_(y, X_row)
        dists.append((X_row, dist))
    dists.sort(key = lambda x: x[1])
    nearestn = []
    for i in range(k):
        nearestn.append(dists[i])
    return nearestn
```
